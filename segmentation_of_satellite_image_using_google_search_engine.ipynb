```python
import geemap
import ee

# Initialize Google Earth Engine
ee.Initialize()

# Create an interactive map centered on India
m = geemap.Map(center=[20.5937, 78.9629], zoom=5)

# Load a Landsat 8 satellite image
india = ee.Geometry.Rectangle([68, 6, 97, 37])  # Bounding box of India
image = ee.ImageCollection("LANDSAT/LC08/C02/T1_TOA") \
         .filterBounds(india) \
         .filterDate("2023-01-01", "2023-12-31") \
         .median()  # Take the median image of the year

# Add to map
m.addLayer(image, {'bands': ['B4', 'B3', 'B2'], 'min': 0, 'max': 0.3}, "Landsat Image - India")
m
```

```python
# Select necessary bands for classification
bands = ['B4', 'B3', 'B2', 'B5', 'B6']  # Red, Green, Blue, Near-Infrared, Shortwave-Infrared

# Create a training dataset from the image
training = image.select(bands).sample(region=india, scale=30, numPixels=5000)

# Train an unsupervised K-means classifier
clusterer = ee.Clusterer.wekaKMeans(5).train(training)

# Apply the classifier to segment the image
classified = image.select(bands).cluster(clusterer)

# Define colors for each class
palette = ['blue', 'gray', 'green', 'yellow', 'brown']  # Water, Urban, Vegetation, Barren, Soil

# Add segmented map to the interactive view
m.addLayer(classified.randomVisualizer(), {}, "Segmented Map")
m
```

## now using random forest

```python
import geemap
import ee

# Initialize GEE
ee.Initialize()

# Create an interactive map centered on India
m = geemap.Map(center=[20.5937, 78.9629], zoom=5)

# Load Landsat 8 image (2023)
india = ee.Geometry.Rectangle([68, 6, 97, 37])  # Bounding box of India
image = ee.ImageCollection("LANDSAT/LC08/C02/T1_TOA") \
         .filterBounds(india) \
         .filterDate("2023-01-01", "2023-12-31") \
         .median()

# Select bands useful for classification
bands = ['B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7']  # Coastal, Blue, Green, Red, NIR, SWIR1, SWIR2

# Add RGB image to map
m.addLayer(image, {'bands': ['B4', 'B3', 'B2'], 'min': 0, 'max': 0.3}, "True Color Image")
m
```

```python
# Sample points for different land types
urban = ee.FeatureCollection([
    ee.Feature(ee.Geometry.Point([77.2090, 28.6139]), {'class': 0}),  # New Delhi
    ee.Feature(ee.Geometry.Point([72.8777, 19.0760]), {'class': 0})   # Mumbai
])

water = ee.FeatureCollection([
    ee.Feature(ee.Geometry.Point([78.4767, 23.5859]), {'class': 1}),  # Narmada River
    ee.Feature(ee.Geometry.Point([76.2673, 9.9570]), {'class': 1})   # Kerala Backwaters
])

vegetation = ee.FeatureCollection([
    ee.Feature(ee.Geometry.Point([78.0421, 27.1751]), {'class': 2}),  # Forest near Taj Mahal
    ee.Feature(ee.Geometry.Point([76.8657, 20.3830]), {'class': 2})   # Central India Greenlands
])

barren = ee.FeatureCollection([
    ee.Feature(ee.Geometry.Point([70.9010, 27.0238]), {'class': 3}),  # Rajasthan Desert
    ee.Feature(ee.Geometry.Point([71.8374, 23.2156]), {'class': 3})   # Kutch Barren Land
])

agriculture = ee.FeatureCollection([
    ee.Feature(ee.Geometry.Point([75.7873, 26.9124]), {'class': 4}),  # Punjab Farmlands
    ee.Feature(ee.Geometry.Point([74.3587, 31.5204]), {'class': 4})   # Haryana Croplands
])

# Merge all training data
training_data = urban.merge(water).merge(vegetation).merge(barren).merge(agriculture)
```

```python
# Extract values of selected bands from the image for training
training = image.select(bands).sampleRegions(
    collection=training_data,
    properties=['class'],
    scale=30
)

# Train the Random Forest model
classifier = ee.Classifier.smileRandomForest(10).train(training, 'class', bands)

# Classify the entire image
classified = image.select(bands).classify(classifier)

# Define a color palette for each class
palette = ['gray', 'blue', 'green', 'yellow', 'brown']  # Urban, Water, Vegetation, Barren, Agriculture

# Add the classified map
m.addLayer(classified, {'min': 0, 'max': 4, 'palette': palette}, "Land Classification")
m
```

```python
import geemap
import ee

# Initialize Google Earth Engine
ee.Initialize()

# Create an interactive map centered on Delhi
m = geemap.Map(center=[28.6139, 77.2090], zoom=10)

# Define Delhi boundary
delhi = ee.Geometry.Rectangle([76.85, 28.4, 77.3, 28.9])  # Approximate bounding box

# Load Sentinel-2 image collection (High-Resolution)
image = ee.ImageCollection("COPERNICUS/S2") \
         .filterBounds(delhi) \
         .filterDate("2023-01-01", "2023-12-31") \
         .median()  # Taking median to remove clouds

# Select high-resolution bands (10m resolution)
bands = ['B2', 'B3', 'B4', 'B8']  # Blue, Green, Red, Near-Infrared

# Add True Color Image to Map
m.addLayer(image, {'bands': ['B4', 'B3', 'B2'], 'min': 0, 'max': 3000}, "Sentinel-2 Delhi")
m
```

```python
# Define training data points for different land types
urban = ee.FeatureCollection([
    ee.Feature(ee.Geometry.Point([77.2167, 28.6448]), {'class': 0}),  # Connaught Place
    ee.Feature(ee.Geometry.Point([77.1987, 28.5661]), {'class': 0})   # South Extension
])

water = ee.FeatureCollection([
    ee.Feature(ee.Geometry.Point([77.2375, 28.6358]), {'class': 1}),  # Yamuna River
    ee.Feature(ee.Geometry.Point([77.2712, 28.6097]), {'class': 1})   # Okhla Barrage
])

vegetation = ee.FeatureCollection([
    ee.Feature(ee.Geometry.Point([77.2160, 28.6120]), {'class': 2}),  # Lodhi Garden
    ee.Feature(ee.Geometry.Point([77.2312, 28.6333]), {'class': 2})   # Central Ridge Forest
])

roads = ee.FeatureCollection([
    ee.Feature(ee.Geometry.Point([77.2106, 28.6341]), {'class': 3}),  # Rajpath
    ee.Feature(ee.Geometry.Point([77.1722, 28.5244]), {'class': 3})   # NH-48 Highway
])

bare_soil = ee.FeatureCollection([
    ee.Feature(ee.Geometry.Point([77.0584, 28.5746]), {'class': 4}),  # Dry open land near Gurugram
    ee.Feature(ee.Geometry.Point([77.1035, 28.6928]), {'class': 4})   # Near Rohini
])

# Merge all training data
training_data = urban.merge(water).merge(vegetation).merge(roads).merge(bare_soil)
```

```python
# Extract values for training
training = image.select(bands).sampleRegions(
    collection=training_data,
    properties=['class'],
    scale=10  # Sentinel-2 has 10m resolution
)

# Train a high-precision Random Forest model (50 trees for better accuracy)
classifier = ee.Classifier.smileRandomForest(50).train(training, 'class', bands)

# Apply classification
classified = image.select(bands).classify(classifier)

# Define a color palette for each class
palette = ['gray', 'blue', 'green', 'black', 'yellow']  # Urban, Water, Vegetation, Roads, Bare Soil

# Add the classified map
m.addLayer(classified, {'min': 0, 'max': 4, 'palette': palette}, "High-Precision Land Classification")
m
```
